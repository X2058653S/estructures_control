```python
#Exercici 4
#Crea un programa que donada una llista qualsevol, et digui si és simètrica o no. 
#Si ho és, que et digui quants elements té.

import random

#Creamos una lista random

list=(random.sample(range(20),7))
print('lista original :',list)

#Creamos dos listas separando la lista original por la mitad, 
#en función de si su cantidad de elementos es par o impar:

n=len(list)
m=(len(list)//2)

if n%2>0:
    list1=list[:m]
    list2=list[(m+1):]
else:
    list1=list[:m]
    list2=list[m:]

print('lista 1ª :', list1)
print('lista 2ª :', list2)

#Comparamos las 2 listas obtenidas:

if list1==list2:
    print('la lista original es simétrica i té ',n,' elements')
else:print('la lista original no es simétrica') 

```

```python
#Exercici 5
#Crea un programa que donada una llista, et digui quants números coincideixen amb la seva posició. 
#Per exemple [3,4,2,0,2,3,6] el 2 i el 6 coincideixen.

#Creamos una lista random:

llista = (random.sample(range(20),20))
print('mi lista es:',llista)

#Creamos una lista con las posiciones de los elementos de la lista primera:

llista2=[]
for x in range(len(llista)):
    llista2.append(x)


#Creamos una lista resultante de la resta de las dos listas, 
#si el resultado es 0, el número coincide con su posición:

llista3=[]
for y in range(len(llista2)):
    llista3.append(llista[y]-llista2[y])


#Creamos una lista con los elementos que coinciden con su posición:
llista4=[a for a,b in enumerate(llista3) if b==0]
print('los números de la lista que coinciden con su posición son:',llista4)

#Contamos la cantidad de elementos que coinciden con su posición:
print('la cantidad de números que coincide con su posición es:',len(llista4))



```

```python
#Vamos a crear un caso en el cual la lista tiene números que coinciden con su posición
#para verificar validez del programa:

llista = (0,1,2,10,11,5,13)
print('mi lista es:',llista)

llista2=[]
for x in range(len(llista)):
    llista2.append(x)

llista3=[]
for y in range(len(llista2)):
    llista3.append(llista[y]-llista2[y])

llista4=[a for a,b in enumerate(llista3) if b==0]
print('los números de la lista que coinciden con su posición son:',llista4)

print('la cantidad de números que coincide con su posición es:',len(llista4))
```
